# deskripsi library yang digunakan

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Memanggil Dataset

```python
df = pd.read_csv('Pizza.csv')
```

# Deskripsi Dataset

```python
df.head()
```

```python
df.info()
```

```python
sns.heatmap(df.isnull())
```

```python
df.describe()
```

# Visualisasi Data

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(),annot=True)
```

# Jumlah Kalori berdasarkan brand

```python
models = df.groupby('brand').count()[['prot']].sort_values(by='prot',ascending=True).reset_index()
models = models.rename(columns={'prot':'KaloryofCategory'})
```

```python
fig = plt.figure(figsize=(15,5))
sns.barplot(x=models['brand'], y=models['KaloryofCategory'], color='royalblue')
plt.xticks(rotation=60)
```

# distribusi Jumlah Kalori

```python
plt.figure(figsize=(15,5))
sns.distplot(df['cal'])
```

# seleksi fitur

```python
features = ['prot','carb','sodium','fat']
x = df[features]
y = df['cal']
x.shape, y.shape
```

# split data training dan data testing

```python
from sklearn.model_selection import train_test_split
x_train, X_test, y_train, y_test = train_test_split(x,y,random_state=70)
y_test.shape
```

# membuat model regresi linier

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
pred = lr.predict(X_test)
```

```python
score = lr.score(X_test, y_test)
print('akurasi model regresi linier = ', score)
```

# membuat inputan model regresi linier

```python
# prot:19.99,carb:0.80,sodium:1.63,fat:45.78
input_data = np.array([[19.99,0.80,1.63,45.78]])

prediction = lr.predict(input_data)
print('Estimasi jumlah kalori pada pizza :' ,prediction ,'kcal')
```

# save model

```python
import pickle

filename = 'Pizza.sav'
pickle.dump(lr,open(filename,'wb'))
```
